```python
import random
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np

def cordenX():
  deltaX = random.randint(0,1000)
  return deltaX/10

def cordenY():
  deltaY = random.randint(0,1000)
  return deltaY/10

def eficiencia():
  roll= random.randint(1,100)
  if roll <= 50:
    return True
  else:
    return False

numparticulas=1000
coordenadaenX=[]
coordenadaenY=[]
parts=0
detectorenX=[]
detectorenY=[]
detectadoX=[]
detectadoY=[]
notdetectedX=[]
notdetectedY=[]
detectorenX.append(40)
detectorenY.append(40)

detectorenX.append(60)
detectorenY.append(40)

detectorenX.append(60)
detectorenY.append(60)

detectorenX.append(40)
detectorenY.append(60)

detectorenX.append(40)
detectorenY.append(40)



while parts < numparticulas:
  x= cordenX()
  y= cordenY()
  coordenadaenX.append(x)
  coordenadaenY.append(y)
  if(40<x<60) and (40<y<60):
    if eficiencia():
      detectadoX.append(x)
      detectadoY.append(y)
    else:
      notdetectedX.append(x)
      notdetectedY.append(y)
  parts+=1

plt.plot(coordenadaenX,coordenadaenY, 'bo')
plt.plot(detectorenX,detectorenY,'g')
plt.plot(detectadoX, detectadoY,'ro')
plt.plot(notdetectedX, notdetectedY,'yx')
```
