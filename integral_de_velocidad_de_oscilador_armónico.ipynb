#**¿Qué hay que hacer?**

Usar la sumatoria de Riemann para aproximarnos a $\int_{a}^{b}Awcos(wt)dt$ en un intervalo $t\in[a,b]$

Usamos entonces, $\sum_{i=1}^{n}{f(a+\frac{i(b-a)}{n})(\frac{b-a}{n}})$ donde $\frac{b-a}{n}$ es la partición 

$f(a+\frac{i(b-a)}{n})$ = $Awcos(a + \frac{i(b-a)}{n})$

Entonces 

$\sum_{i=1}^{n}{f(a+\frac{i(b-a)}{n})(\frac{b-a}{n}})$ = $\sum_{i=1}^{n}Awcos(a + \frac{i(b-a)}{n})(\frac{b-a}{n})$



Tomamos ciertos valores como ejemplo, tales que

$A = 2, P = 2, a = 2, b = 5, n = 100$.

Donde **A** es la amplitud del oscilador, **P** el periodo, **a** el punto inicial, **b** el punto final y **n** el número de particiones



```python
import numpy as np
def function1(t): 
  return A*(2*np.pi/P)*np.cos((2*np.pi/P)*t)
def function2(t):
  return A*np.sin((2*np.pi/P)*t)
print ("¿Cuál es la amplitud del oscilador?")
#w= (2*np.pi/P) #rapidez angular 
A= float (input(""))
print ("¿Cuál es el periodo del oscilador en segundos?")
P = float (input("")) 


def sumariemann(function1,a,b,n):
    sumval = 0
    h = (b-a)/n
    sumval = h * sum(function1(a + i * h) for i in range(0,n) )
    return sumval 

print ("¿Desde qué punto quieres calcular la integral?") #intervalo [a,b]
a= int(input(""))
print ("¿Hasta qué punto quieres calcular la integral?")
b= int(input(""))
print ("Ingrese número de particiones")
n= int(input(""))
print("La suma de riemann es aproximada a: ",sumariemann(function1,a,b,n))
integralexacta= function2(b)- function2(a)
print("El valor de la integral exacta es: ", integralexacta)
```
