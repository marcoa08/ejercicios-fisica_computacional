```python
#Aguilar Cárcamo Marco Antonio
#Fecha 27/09/21
#Programa de primera y segunda derivada del oscilador armónico con gráficas
import numpy as np
import matplotlib.pyplot as plt

#Definir variables
def function1(t):#funcion
  return np.sin(np.sqrt(k/m)*t)
#t1 = np.arange(1,3,.1)
def function2(t):#primera derivada
  return (function1(t+h) - function1(t))/h
#t1 = np.arange(1,3,.1)
def function3(t):
   return (function2(t+h) - function2(t))/h

#Iniciar variables
print ('Inserte la delta de cambio h')
h= float (input(""))
print ('Inserte la constante k del resorte[N/m]')
k = float (input(""))
print ('Inserte la masa del objeto [kg]')
m = float (input(""))
print ('Inserte t deseado [s]')
t = float (input(""))
t1 = np.arange(1.00,3.00, .01)

#Impresión de datos numéricos
print('El valor numérico de la función en t es', function1(t), 'm')
#print('El valor numérico aproximado de la primera derivada en t es ', (function1(t+h) - function1(t))/h, 'm/s')
print('El valor numérico aproximado de la primera derivada en t es ', function2(t), 'm/s')
#print('El valor numérico aproximado de la segunda derivada en t es ', (function2(t+h) - function2(t))/h, 'm/s^2')
print('El valor numérico aproximado de la segunda derivada en t es ', function3(t), 'm/s^2')

#Impresión de gráficas
plt.plot(t1, function1(t1))
plt.title("gráfica de $f(x)$")
plt.show()
plt.plot(t1, function2(t1))
plt.title("gráfica de $f'(x)$")
plt.show()
plt.plot(t1, function3(t1))
plt.title("gráfica de $f''(x)$")
plt.show()




```
